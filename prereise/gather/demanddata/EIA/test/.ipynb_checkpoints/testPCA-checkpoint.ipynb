```python
data = pd.read_csv('this.csv', index_col='Date', parse_dates =True)
```

```python
type(data.index)
```

```python
data.columns = ['West', 'East']
data.columns = ['West', 'East']
data['Total'] = data.eval('West + East')
data.resample('W').sum().plot()
```

```python
pivoted = data.pivot_table('Total', index=data.index.time, columns = data.index.date)
pivoted.plot(legend=False, alpha =0.01)
```

```python
#Convert to numpy
X = pivoted.fillna(0).T.values
X.shape
```

```python
from sklearn.decomposition import PCA
X2 = PCA(n_components = 2, svd_solver='full').fit_transform(X)
X2.shape
```

```python
plt.scatter(X2[:,0],X2[:,1])
```

```python
# What are clusters
from sklearn.mixture import GaussianMixture
gmm = GaussianMixture(2)
gmm.fit(X)
labels = gmm.predict(X)
labels
```

```python
plt.scatter(X2[:,0],X2[:,1], c=labels, cmap = 'rainbow')
plt.colorbar()
```

```python
pivoted.T[labels == 0].T.plot(legend=False, alpha = 0.1)
```

```python
pivoted.T[labels == 1].T.plot(legend=False, alpha = 0.1)
```
