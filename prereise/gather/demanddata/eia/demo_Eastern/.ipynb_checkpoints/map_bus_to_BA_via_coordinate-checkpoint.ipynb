```python
import os
import fiona
import rtree
import shapely.ops
import pyproj
import functools
from shapely.geometry import shape, box, Polygon, MultiPolygon
from tqdm import tqdm_notebook as tqdm
```

### Read the shape file of control areas from HIFLD database via Fiona package. Build a rTree based on the shape file storing boundaries of each polygon and all the properties in different nodes for fast locating a polygon (point) to the nearest polygon in the shape file. Note that the polygons may overlap with each other.

```python
def build_rtree_index(shape_file_dir, rtree_file_dir, utm_zone, bbox=None, overwrite=False):
    """
    Build an R-Tree for the given shapefile based on given bounding box, store the polygons with attributes
    Arguments:
        shape_file_dir {str} -- Directory of the input shapefle
        rtree_file_dir {str} -- Directory to save R-Tree file to this directory
        utm_zone {int} -- UTM zone used to convert to UTM coordinate
    Keyword Arguments:
        bbox {list[float]} -- If not None, only keep polygons within bbox (default: {None}) bbox format: [left, bottom, right, top]
        overwrite {bool} -- Whether to overwrite existing files (default: {False})
    Returns:
        [rtree.index.Index] -- The R-Tree index built based on shapefile
    """
    # Clear old file
    if os.path.exists(rtree_file_dir + '.idx'):
        if overwrite:
            print(f'{rtree_file_dir} already exists, clear the old R-Tree files and rebuild')
            os.remove(rtree_file_dir + '.idx')
            os.remove(rtree_file_dir + '.dat')
        else:
            print(f'{rtree_file_dir} already exists, use the pre-calculated R-Tree directly')
            return rtree.index.Index(rtree_file_dir)
    # Open shapefile and R-Tree index
    coll = fiona.open(shape_file_dir)
    rtree_idx = rtree.index.Index(rtree_file_dir)
    # Create transformation function for coordinate conversion
    # Doing a dumb transformation here. Since the HIFLD database metadata indicates they are using EPSG:4326 coordination
    # system, which is also the standard reference of the GPS satellite navigation system
    proj_in = pyproj.Proj(init='epsg:4326')
#     proj_out = pyproj.Proj(init=f'epsg:326{utm_zone}')
    proj_out = pyproj.Proj(init='epsg:4326')
    project = functools.partial(pyproj.transform, proj_in, proj_out)
    # Create iterator over interest polygons
    total = sum(1 for _ in (coll.items(bbox=bbox) if bbox else iter(coll)))
    with tqdm(total=total) as pbar:
        it = coll.items(bbox=bbox) if bbox else iter(coll)
        invalid_count = 0
        for i, item in enumerate(it):
            # Validate polygons
            p = shape(item['geometry'])
            if not p.is_valid:
                p = p.buffer(0, 0)
            if len(p.bounds) == 0:
                invalid_count += 1
                continue
            # Cast geometry record into shapely object
            poly = shapely.ops.transform(project, p)
            record = list(item['properties'].values())
            # Insert new item to R-Tree
            rtree_idx.insert(i, poly.bounds, obj=(poly, record))
            # Update progress bar
            if i % 100 == 0:
                pbar.set_postfix(invalid_count=invalid_count)
                pbar.update(100)
    # Close file to save to disk first and then reopen
    rtree_idx.close()
    rtree_idx = rtree.index.Index(rtree_file_dir)
    return rtree_idx
```

```python
r_idx = build_rtree_index('./Control_Areas/Control_Areas.shp','./Contral_Areas',None, overwrite = True)
```
