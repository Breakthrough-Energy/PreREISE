### Find County names that have a different spelling than the plotting tool

```python
data = json.load(open('MISOAllStateCounties.txt'))
```

```python
tool_MISO_counties = set(data['groups']['#cc3333']['paths'])
```

```python
MISO[3] = MISO[3].apply(lambda row : row.replace(' ','_'))
```

```python
MISO[3] = MISO[3].apply(lambda row : row.replace('.',''))
```

```python
MISO.loc[MISO[3] == "O'Brien__IA",3] = 'O_Brien__IA'
```

```python
MISO.loc[MISO[3] == "St_Louis_City__MO",3] = 'St_Louis__MO'
```

```python
MISO.loc[MISO[3] == "La_Salle__LA",3] = 'LaSalle__LA'
```

```python
MISO.loc[MISO[3] == "St_Louis_County__MO",3] = 'St_Louis_Co__MO'
```

```python
MISO.loc[MISO[3] == "LaSalle__IL",3] = 'La_Salle__IL'
```

```python
yixing_MISO_counties = set(list(MISO[3].values))
```

```python
counties_misspelled = (yixing_MISO_counties^(tool_MISO_counties)).intersection(yixing_MISO_counties)
```

```python
import difflib
```

```python
for county in counties_misspelled:
    print(county ,difflib.get_close_matches(county, tool_MISO_counties))
    
```

### Find overlaping counties

```python
data = json.load(open('EasternBAtoCountyDraft6.txt'))
```

```python
ba_sets = []
ba_name = []
for ba in data['groups']:
    ba_name.append(data['groups'][ba]['label'])
    ba_sets.append(set(data['groups'][ba]['paths']))
```

```python
for i,ba_set in enumerate(ba_sets):
    print(ba_name[i])
    for j in range(i,len(ba_sets)):
        if i==j:
            continue
        print(' ',ba_name[j],ba_set.intersection(ba_sets[j]))
```
