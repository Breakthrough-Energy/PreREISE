```python
bus_map = pd.read_csv('bus_ba_map.csv')
bus_map.set_index('bus_id')
```

```python
bus_map['BA'].unique()
```

```python
BA_agg = bus_map[['BA','Pd']].groupby('BA').sum()
BA_agg
```

```python
BA_agg.loc['AEC','Pd']
```

```python
bus_map.BA.apply(lambda x: BA_agg.loc[x,'Pd'])
```

```python
BA_scaling_factor = bus_map.merge(BA_agg, left_on='BA', right_on='BA')
BA_scaling_factor
```

```python
BA_scaling_factor = BA_scaling_factor.assign(zone_scaling = lambda x: x['Pd_x']/x['Pd_y'])
BA_scaling_factor
```

```python
BA_scaling_factor
```

```python
BA_grouped_scaling = BA_scaling_factor.groupby(['BA','zone_name']).sum()
BA_grouped_scaling
```

```python
eastern_agg_demand = pd.read_pickle("C:\\Users\\dmuldrew\\Dropbox (IVL)\\Explorations\\DanM\\demand_interpolated.pkl")
eastern_agg_demand
```

```python
eastern_zone_demand = pd.DataFrame(index=eastern_agg_demand.index)

for ba_name in BA_grouped_scaling.index.unique(level = 'BA').to_list():
    
    zone_scaling_df = BA_grouped_scaling.loc[ba_name]
    print(zone_scaling_df['zone_scaling'])
    print(zone_scaling_df['zone_scaling'].sum())
    print(zone_scaling_df.index.get_level_values(0).to_list())
    print()
    for zone_name in zone_scaling_df.index.get_level_values(0).to_list():
        print(ba_name)
        print(zone_name)
        print(zone_scaling_df.loc[zone_name,'zone_scaling'])
        print()
        if zone_name in eastern_zone_demand.columns:
            eastern_zone_demand[zone_name] += (zone_scaling_df.loc[zone_name,'zone_scaling']*Eastern_agg_demand[ba_name])
        else:
            eastern_zone_demand[zone_name] = (zone_scaling_df.loc[zone_name,'zone_scaling']*Eastern_agg_demand[ba_name])
```
