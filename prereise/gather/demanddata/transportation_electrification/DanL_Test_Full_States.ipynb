```python
### Running LA in 2035 for both Immediate and Smart

import os

import numpy as np
import pandas as pd
from scipy.io import loadmat

from prereise.gather.demanddata.transportation_electrification import (
    const,
    immediate,
    immediate_charging_HDV,
    smart_charging,
)
from prereise.gather.demanddata.transportation_electrification.data_helper import (
    generate_daily_weighting,
    get_kwhmi,
    load_rural_scaling_factor,
    load_urbanized_scaling_factor,
)


#data_dir = os.path.join(
#    const.data_folder_path,
#    "CAISO_sample_load_2019.mat",
#)
#load_demand = loadmat(data_dir)["load_demand"].flatten()

# LOAD NEW NET DEMAND
data_dir = os.path.join(
    const.data_folder_path,
    "LA_external_signal_2035.csv",
)
load_demand = pd.read_csv(
    data_dir #, index_col="Value"
)



#charging_strategy="immediate"
charging_strategy="smart"
veh_type=["mdv","hdv","ldv","ldv","ldv","ldt","ldt","ldt"]
veh_range=[200,200,100,200,300,100,200,300] 
projection_year=2035
state="CA"
location_strategy=[1,1,2,2,2,2,2,2]
trip_strategy=1
external_signal=load_demand["Value"]

census_region = const.state2census_region[state]

daily_values = generate_daily_weighting(projection_year)

urban_bev_vmt = [1179462692,965497208.9,4829168945,3148369490,3322828566,8432410367,4383994989,369072249.4]

#Los Angeles-Long Beach-Anaheim, CA	CA	Los Angeles-Long Beach-Anaheim	
#urban_bev_vmt = 4829168945	#LDV-100
#urban_bev_vmt = 3148369490	#LDV-200
#urban_bev_vmt = 3322828566	#LDV-300
#urban_bev_vmt = 8432410367	#LDT-100
#urban_bev_vmt = 4383994989	#LDT-200
#urban_bev_vmt = 369072249.4   #LDT-300
#urban_bev_vmt = 1179462692	#MDV
#urban_bev_vmt = 965497208.9	#HDV

LA_2035_demand_imm_profiles_df = pd.DataFrame()
LA_2035_demand_smart_profiles_df = pd.DataFrame()
state_demand_imm_profiles = {}

#Loop over vehicle types

for i in range(len(veh_type)):
#for i in range(2):

    print(f"On Vehicle {veh_type[i]} with range {veh_range[i]} and BEV VMT of {urban_bev_vmt[i]}")
    # calculate demand for all geographic areas with scaling factors
    #state_demand_smart_profiles = {}

    kwhmi = get_kwhmi(projection_year, veh_type[i], veh_range[i])

    if veh_type[i].lower() in {"ldv", "ldt"}:
        vehicle_trip_data_filepath=os.path.join(
            const.data_folder_path,
            "nhts_census_updated_dwell.mat",
        )
        power=6.6        
        if power > 19.2:
            charging_efficiency = 0.95
        else:
            charging_efficiency = 0.9

        if charging_strategy == "immediate":
            normalized_demand, _, _ = immediate.immediate_charging(
                census_region=census_region,
                model_year=projection_year,
                veh_range=veh_range[i],
                power=power,
                location_strategy=location_strategy[i],
                veh_type=veh_type[i],
                filepath=vehicle_trip_data_filepath,
            )
        elif charging_strategy == "smart":
            final_smart_demand, _, _ = smart_charging.smart_charging(
                census_region=census_region,
                model_year=projection_year,
                veh_range=veh_range[i],
                kwhmi=kwhmi,
                power=power,
                location_strategy=location_strategy[i],
                veh_type=veh_type[i],
                filepath=vehicle_trip_data_filepath,
                daily_values=daily_values,
                external_signal=external_signal,
                bev_vmt=urban_bev_vmt[i],
                trip_strategy=trip_strategy,
            )


    elif veh_type[i].lower() in {"mdv", "hdv"}:
        #HDV
        vehicle_trip_data_filepath=os.path.join(
            const.data_folder_path,
            "fdata_v10st.mat",
        )
        power=80        
        if power > 19.2:
            charging_efficiency = 0.95
        else:
            charging_efficiency = 0.9

        if charging_strategy == "immediate":
            normalized_demand, _, _ = immediate_charging_HDV.immediate_hdv_charging(
                model_year=projection_year,
                veh_range=veh_range[i],
                power=power,
                location_strategy=location_strategy[i],
                veh_type=veh_type[i],
                filepath=vehicle_trip_data_filepath,
            )
        elif charging_strategy == "smart":
            final_smart_demand, _, _ = smart_charging.smart_charging(
                census_region=census_region,
                model_year=projection_year,
                veh_range=veh_range[i],
                kwhmi=kwhmi,
                power=power,
                location_strategy=location_strategy[i],
                veh_type=veh_type[i],
                filepath=vehicle_trip_data_filepath,
                daily_values=daily_values,
                external_signal=external_signal,
                bev_vmt=urban_bev_vmt[i],
                trip_strategy=trip_strategy,
            )


    if charging_strategy == "immediate":
        state_demand_imm_profiles = immediate.adjust_bev(
            hourly_profile=normalized_demand,
            adjustment_values=daily_values,
            model_year=projection_year,
            veh_type=veh_type[i],
            veh_range=veh_range[i],
            bev_vmt=urban_bev_vmt[i],
            charging_efficiency=charging_efficiency,
        )/1000

        external_signal += state_demand_imm_profiles

        temp_result = pd.DataFrame(
            state_demand_imm_profiles,
            index=pd.date_range(
                start=f"{projection_year}-01-01 00:00:00",
                end=f"{projection_year}-12-31 23:00:00",
                freq="H",
            ),
        )

        LA_2035_demand_imm_profiles_df = LA_2035_demand_imm_profiles_df.append(temp_result, ignore_index=True)
    
    elif charging_strategy == "smart":
        external_signal += final_smart_demand

        smart_result = pd.DataFrame(
            final_smart_demand,
            index=pd.date_range(
                start=f"{projection_year}-01-01 00:00:00",
                end=f"{projection_year}-12-31 23:00:00",
                freq="H",
            ),
        )

        LA_2035_demand_smart_profiles_df = LA_2035_demand_smart_profiles_df.append(smart_result, ignore_index=True)

```
